```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcol
import matplotlib.ticker as ticker

import seaborn as sns
import numpy as np
import pandas as pd
import os
os.chdir('/project2/jnovembre/ccliu/backgroundSelection')
```

```python
N = 5000
mu = 6.25 * 1e-7
l = 5e5
Ud = mu * l * 4/25

s = -0.1
B = np.exp(Ud / s * 2)
print(f'B =  {B}')
print(f'Ne x s =  {- N * B * s}')
```

## Three code chunks for creating figures for WIP presentation
1. Creating a job list (Then go to terminal and run script/infer_gamma_3classes.sh)
2. Read in outputs
3. Plot

```python
reps = range(1, 21)
out = f'../output/gamma_inference/bgs_3classes_N5000.c.txt'

with open('script/file_list.txt', 'w') as f:
    for rep in reps:
        for c in range(30):
            file = f'../data/slim/bgs_3classes/bgs_3classes_N5000_rep{rep}.trees'
            f.write(f'{file} 500 {c} {out}\n')
```

```python
inferred_gammas = []
with open('output/gamma_inference/bgs_3classes_N5000.n.txt') as f:
    for line in f:
        tree, n, c, gamma = line.strip().split()
        #if int(n) >= 15:
        inferred_gammas.append((int(n), float(gamma)))
inferred_gammas = pd.DataFrame(inferred_gammas, columns =['sample size', 'inferred gamma']) 
```

```python
ax = sns.boxplot(x="sample size", y="inferred gamma", data=inferred_gammas, color = 'grey',
                 showmeans=True,
                 meanprops={"marker":"o",
                            "markerfacecolor":"red", 
                            "markeredgecolor":"black",
                            "markersize":"3"})
#ax.xaxis.set_major_locator(ticker.LinearLocator(5))
plt.axhline(y = N * -0.02 * B , color='blue', linestyle='-.')
#ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
#ax.xaxis.set_ticks(np.arange(35,500,50))
plt.xlabel("sample size", fontsize = 15)
plt.ylabel("inferred gamma", fontsize = 15)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    
plt.savefig('figs/inferred_gamma_n.png', dpi = 600, bbox_inches='tight')
```

```python
# load results
inferred_gammas = []
with open('output/gamma_inference/bgs_3classes_N5000.c.txt') as f:
    for line in f:
        tree, n, c, gamma = line.strip().split()
        if int(c) <= 25:
            inferred_gammas.append((int(c) * 2, float(gamma)))
inferred_gammas = pd.DataFrame(inferred_gammas, columns =['Number of entries truncated', 'Inferred gamma']) 
```

```python
ax = sns.boxplot(x="Number of entries truncated",
                 y="Inferred gamma", 
                 data=inferred_gammas,
                 color = 'grey',
                 showmeans=True,
                 meanprops={"marker":"o",
                            "markerfacecolor":"red", 
                            "markeredgecolor":"black",
                            "markersize":"3"})
#ax.xaxis.set_major_locator(ticker.LinearLocator(5))
plt.axhline(y = N * -0.02 * B , color='blue', linestyle='-.')
#ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
#ax.xaxis.set_ticks(np.arange(0,30,2))
plt.xlabel("Number of entries truncated", fontsize = 15)
plt.ylabel("Inferred gamma", fontsize = 15)
plt.ylim(-150, 0)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    
plt.savefig('figs/inferred_gamma_c.png', dpi = 600, bbox_inches='tight')
```
